```python
import numpy as np
import matplotlib.pyplot as plt

# Example data for stress (MPa) and strain (unitless) from a tensile test
strain = np.linspace(0, 0.01, 100)  # Strain from 0 to 1% strain
# Stress increases linearly in the elastic region and then flattens out
stress = 250 * strain  # Example: Linear relationship for the elastic region

# 0.2% offset strain (0.002)
offset_strain = 0.002
# Calculate the slope of the initial linear region (Young's Modulus)
modulus = 250  # Example Young's Modulus in MPa (slope of the linear region)
# Draw the 0.2% offset line (parallel to the elastic region but shifted)
offset_stress = modulus * (strain + offset_strain)

# Plotting the stress-strain curve
plt.plot(strain, stress, label='Stress-Strain Curve', color='blue', linewidth=2)
plt.plot(strain, offset_stress, label='0.2% Offset Line', color='red', linestyle='--')

# Find the intersection point (yield strength)
yield_point_index = np.argmax(offset_stress >= stress)  # Find where offset line intersects curve
yield_strain = strain[yield_point_index]
yield_stress = stress[yield_point_index]

# Mark the yield point
plt.plot(yield_strain, yield_stress, 'go', label=f'Yield Point ({yield_strain:.4f}, {yield_stress:.2f} MPa)')

# Adding labels and title
plt.xlabel('Strain')
plt.ylabel('Stress (MPa)')
plt.title('Stress-Strain Curve with 0.2% Offset Yield')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('temp/500.csv')

strain_rates = data['strain_rate'].unique()


colors = plt.cm.viridis(np.linspace(0, 1, len(strain_rates)))

# Creating the plot
plt.figure(figsize=(10, 6))

for i, strain_rate in enumerate(strain_rates):
    subset = data[data['strain_rate'] == strain_rate]
    plt.plot(subset['strain'], subset['stress'],label=f'Strain Rate: {strain_rate:.2e}', color=colors[i])

plt.text(0.95, 0.95, '500K', horizontalalignment='right', verticalalignment='top', 
         transform=plt.gca().transAxes, fontsize=13,)
# Adding legneds and titles
plt.title('Stress-Strain Curve at Different Strain Rates')
plt.xlabel('Strain')
plt.ylabel('Stress')
plt.legend()
plt.grid()
plt.show()

```
